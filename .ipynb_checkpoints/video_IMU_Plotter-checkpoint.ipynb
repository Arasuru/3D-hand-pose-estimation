```python
import cv2
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.use('TkAgg')
```

```python
video_path = r"C:\Users\ravik\practice\projects\Hand Pose detection\mediapipe videos\OBS recordings\2025-02-24 11-44-44.mp4"
cap = cv2.VideoCapture(video_path)
```

```python
data = pd.read_csv(r"C:\Users\ravik\Desktop\Thesis\Glove\recordings\trial1_full_use.csv")
data.head()
```

```python
sensor_dfs = {sensor: group for sensor, group in data.groupby(' Sensor')}

# Print each sensor's DataFrame
for sensor, sensor_df in sensor_dfs.items():
    print(f"Sensor {sensor} DataFrame:")
    print(sensor_df.head())
    print()
```

```python
def plot_imu_data(imu_data, ax):
    ax.plot(imu_data['Timestamp'], imu_data[' QuatI'], label='Accel X')
    ax.plot(imu_data['Timestamp'], imu_data[' QuatJ'], label='Accel Y')
    ax.plot(imu_data['Timestamp'], imu_data[' QuatK'], label='Accel Z')
    ax.plot(imu_data['Timestamp'], imu_data[' QuatSum'], label='Accel')
    ax.set_xlabel('Timestamp')
    ax.set_ylabel('Acceleration')
    ax.legend()
    ax.set_title("IMU data")
    ax.grid(True)
```

```python
fig, (ax1,ax2,ax3,ax4,ax5,ax6) = plt.subplots(3,2, figsize=(12,6))
ax1.clear()
ax2.clear()
ax3.clear()
ax4.clear()
ax5.clear()
ax6.clear()
plot_imu_data(sensor_dfs[0], ax1)
plot_imu_data(sensor_dfs[1], ax2)
plot_imu_data(sensor_dfs[2], ax3)
plot_imu_data(sensor_dfs[3], ax4)
plot_imu_data(sensor_dfs[4], ax5)
plot_imu_data(sensor_dfs[5], ax6)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
plt.ion()

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    ax1.clear()
    ax2.clear()

    ax1.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
    ax1.set_title("Video feed")
    ax1.axis('off')

    plot_imu_data(sensor_dfs[0], ax2)

    ax2.relim()
    ax2.autoscale_view()

    plt.draw()
    plt.pause(0.001)
    

cap.release()
plt.ioff()
plt.show()
```
